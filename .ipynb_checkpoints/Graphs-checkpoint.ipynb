```python
# Sort the DataFrame by 'MEDIAN_HOUSEHOLD_INCOME' in ascending order
sorted_df = crime_census_df.sort_values(by='MEDIAN_HOUSEHOLD_INCOME')

# Create a bar chart
plt.figure(figsize=(12, 8))
plt.bar(sorted_df['COUNTY_NAME'], sorted_df['MEDIAN_HOUSEHOLD_INCOME'])
plt.title('Total Median Household Income by County')
plt.xlabel('County Name')
plt.ylabel('Median Household Income')
plt.xticks(rotation=90) 
plt.tight_layout()

#style chart
plt.style.use('fivethirtyeight')
plt.show()
```
