```python
import pandas as pd
```

```python
df = pd.read_csv('Data Files/census_data.csv')
```

```python
df
```

```python
#remove NJ from county name
df['COUNTY_NAME'] = df['COUNTY_NAME'].str.replace(', New Jersey', '')

df.head()  
```

```python
#remove county from county name
df['COUNTY_NAME'] = df['COUNTY_NAME'].str.replace(', County', '')

df.head()  
```

```python
# # Write DataFrame to a CSV file
df.to_csv('Data Files/cleaned_census_data.csv', index=False)
```
