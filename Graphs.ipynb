```python
# bar chart comparing how crimes through 2018-2020

# Select values for the bar graph
crime_df = crime_census_df[['COUNTY_NAME','total_2018', 'total_2019', 'total_2020']]
# Plot the bar graph
crime_df.plot(x="COUNTY_NAME", y=['total_2018', 'total_2019', 'total_2020'], kind='bar', figsize=(10, 6), colormap='viridis',rot=0).legend(loc='upper right')
# Create a bar chart
plt.title('Crime 2018 to 2020')
plt.xlabel('County Name')
plt.ylabel('Total Incidents')
plt.xticks(rotation=90)
plt.tight_layout()
#style chart
plt.style.use('fivethirtyeight')
plt.show()
```

```python
# Define the years and corresponding crime categories (removing larcney since it was so much larger)
years = ['2018', '2019', '2020']
crime_categories = ['murder', 'rape', 'robbery', 'assualt',
                    'burglary', 'auto_theft']

# lists to store crime counts for each year
crime_counts_by_year = {year: [] for year in years}

# Populate crime counts for each year and category
for year in years:
    for category in crime_categories:
        crime_column = f'{category}_{year}'
        crime_counts_by_year[year].append(crime_census_df[crime_column].sum())

# Plotting the line graph
plt.figure(figsize=(8, 6))

for category_index, category in enumerate(crime_categories):
    plt.plot(years, [counts[category_index] for counts in crime_counts_by_year.values()], label=category)

plt.xlabel('Year')
plt.ylabel('Total Crime Incidents')
plt.title('Crime Trends by Type (2018-2020)')
plt.legend()
plt.grid(True)
plt.show()

#style chart
plt.style.use('fivethirtyeight')
```
